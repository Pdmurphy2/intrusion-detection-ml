```python
import time
from datetime import datetime

USERNAME = "admin"
PASSWORD = "password123"

login_attempts = 0
failed_attempts = 0

# ⏱️ Start timer immediately when program runs
program_start_time = time.time()
start_readable = datetime.now()
```

```python
def login():
    global login_attempts, failed_attempts
    
    while True:
        username = input("Enter username (or type 'exit'): ")
        
        if username.lower() == "exit":
            print("👋 Exiting program...")
            return "exit"
        
        password = input("Enter password: ")
        
        login_attempts += 1
        
        if username == USERNAME and password == PASSWORD:
            print("✅ Login successful!")
            return "success"
        else:
            failed_attempts += 1
            print("❌ Incorrect credentials.\n")
```

```python
def end_program(status):
    end_time = time.time()
    end_readable = datetime.now()
    
    total_time = end_time - program_start_time
    
    print("\n--- Program Summary ---")
    print(f"Start Time: {start_readable}")
    print(f"End Time:   {end_readable}")
    print(f"Total Time Until {status.upper()}: {total_time:.2f} seconds")
    
    print("\n--- Stats ---")
    print(f"Total login attempts: {login_attempts}")
    print(f"Failed attempts:      {failed_attempts}")
```

```python
result = login()
end_program(result)
```
